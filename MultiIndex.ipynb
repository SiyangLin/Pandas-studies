## The .get_level_values( )Method

```python
#另一种设置多个indexs的方法
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"] )
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

## The sort_index( ) Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"] )
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
#所有的index一起按相同顺序排序
bigmac.sort_index(ascending = False)
```

```python
#按不同的顺序排序
bigmac.sort_index(ascending = [True, False], inplace = True)
```

```python
bigmac.head(3)
```

## The .transpose( ) Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"] )
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
#index变成column
bigmac = bigmac.transpose()
bigmac.head(3)
```

```python
#extract. Notice axis changed
bigmac.loc["Price in US Dollars", ("2015-07-01", "China")]
```

## The .swaplevel ( ) Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"] )
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
#change indexs order,level exchange.改变index的位置顺序
bigmac = bigmac.swaplevel()
bigmac.head(3)
```

## The .stack() Method

## The .unstack() Method, Part 1

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(6)
```

```python
# 返回原表，一步一步拆掉index的个数，从里面往外拆。index move to column 增加维度
s.unstack().unstack()
```

## The .unstack( ) Method. Part 2

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack().to_frame()
s.head()
```

```python
#设置指定的index变成column，不是默认的最里面。输入index的位置数字。如果输入-1.是从右往左数的位置，最后一个column为第一个0.
s.unstack(0)
```

```python
#或者直接输入要移动到列的index名
s.unstack("country")
```

## The .unstack( ) Method, part3

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
s = world.stack().to_frame()
s.head()
```

```python
#将两个index移动到列，同时按分层列，并且指定了层的顺序。
s.unstack(level = [1, 0])
#s.unstack(level = ["year", "country"])
```

```python
#填充 null值
s = s.unstack("year", fill_value = 0)
```

```python
s.head()
```

## The .pivot( ) Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales.head(3)
```

```python
#不同的salesman的数量,很少，5 unique values 可能这几个员工按年工作
sales["Salesman"].value_counts()
```

```python
sales["Salesman"] = sales["Salesman"].astype("category")
```

```python
sales.info()
```

```python
#设置更有意义的横纵坐标
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

## The pivot_table() Method

```python
foods = pd.read_csv("foods.csv")
foods.head()
```

```python
#aggfunc是给予计算式，aggregate 聚集 汇总
foods.pivot_table(values = "Spend", index = "Item", aggfunc = "mean")
```

```python
#group first by gender then by item ,to see how much each gender of each item of the aggregate the total sales总销售额
#aggfunc = "cuont","max"……
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = ["City", "Frequency"], aggfunc = "sum")
```

## THE pd.melt( )Method

```python
sale = pd.read_csv("quarters.csv")
sale
```

```python
#reorient and reorganize the original summarized data frame to a understandable structure
#take column values in single row value,column变成variable，every single value的列也需要命名
pd.melt(sale, id_vars = "Salesman", var_name = "Quarter", value_name = "Revenue")
```

## Bar Charts

```python
bb = data.DataReader(name = "AAPL", data_source = "yahoo", start = "2007-07-01", end = "2008-12-01")
bb.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 15:
        return "Poor"
    elif stock_price > 15 and stock_price <= 20:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
#set bar chart. 如果需要横向的，用kind = "barh"
bb["Close"].apply(rank_performance).value_counts().plot(kind = "bar")
```

## Pie charts

```python
apple = data.DataReader(name = "AAPL", data_source="yahoo", start = "2012-01-01", end = "2016-01-01")
apple.head()
```

```python
apple["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 90.54555943756407:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("ggplot")
```

```python
apple["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True)
```

## Histograms

```python
apple = data.DataReader(name = "AAPL", data_source="yahoo", start = "2012-01-01", end = "2016-01-01")
apple.head()
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
apple["High"].apply(custom_round).value_counts().sort_index()
```

```python
apple["High"].apply(custom_round).nunique()
```

```python
#直方图..........
apple["High"].apply(custom_round).plot(kind = "hist", bins = 2)
```
